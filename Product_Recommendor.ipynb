```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_json('drive/MyDrive/data.json')
```

```python
products = df[["title", "brand", "category", "vendor", "specifications"]]
```

```python
specifications = df["specifications"]
```

```python
specifications_str = specifications.apply(lambda x: " ".join([f"{i[0]}-{i[1]}" for i in x.items()]))
```

```python
products_str = df["title"].astype(str) + " " + df["brand"].astype(str) + " " + df["category"].astype(str) + " " + df["vendor"].astype(str) + " " + specifications_str.astype(str)
```

```python
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(products_str)
```

```python
similarity = cosine_similarity(X)
```

```python
def recommend(i):
  for idx, score in sorted(enumerate(similarity[i]), reverse=True, key=lambda x: x[1])[1:10]:
    print(products.title[idx])
```

```python
recommend(1006)
```
