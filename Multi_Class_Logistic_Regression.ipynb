```python
import numpy as np
```

```python
def sigmoid(x):
  return 1 / (1 + np.exp(-x))
```

```python
def single_prediction(X, weights):
  return sigmoid(np.dot(X, weights))
```

```python
single_prediction([1, 1, 1], [1, 2, 3])
```

```python
def predict(X, weights):
  return sigmoid(np.dot(X, weights))
```

```python
predict([[1, 1, 1], [2, 2, 2]], [1, 2, 3])
```

```python
def logistic_loss(X, y, weights):
  y_pred = predict(X, weights)
  loss = -np.mean(y * np.log(y_pred) + (1 - y) * np.log(1 - y_pred))
  return loss
```

```python
def gradient(X, y, weights):
  m = len(X)
  y_pred = predict(X, weights)
  error = y_pred - y
  gradient = np.dot(X.T, error) / m
  return gradient
```

```python
def gradient_descent(X, y, weights, learning_rate, epochs):
  for epoch in range(epochs):
    gradient_ = gradient(X, y, weights);
    weights -= learning_rate * gradient_
```

```python
def all_predict(X):
  y0 = predict(X, weights_0)
  y1 = predict(X, weights_1)
  y2 = predict(X, weights_2)
  y3 = predict(X, weights_3)
  y4 = predict(X, weights_4)
  y5 = predict(X, weights_5)
  y6 = predict(X, weights_6)
  y7 = predict(X, weights_7)
  y8 = predict(X, weights_8)
  y9 = predict(X, weights_9)
  y_pred = [y0, y1, y2, y3, y4, y5, y6, y7, y8, y9]
  min_ = min(y0, y1, y2, y3, y4, y5, y6, y7, y8, y9)
  return y_pred.index(min_)
```

```python
y_pred = []
for i in range(len(X)):
  y_pred.append(all_predict(X.iloc[i]))
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

accuracy = accuracy_score(y, y_pred)

precision = precision_score(y, y_pred, average='weighted')

recall = recall_score(y, y_pred, average='weighted')

f1 = f1_score(y, y_pred, average='weighted')

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F-1 Score:", f1)
```
