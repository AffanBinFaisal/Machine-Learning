```python
import pandas as pd
```

```python
df = pd.read_json('output.json')
```

```python
def fix_imgs(old_list):
    new_string = str(old_list)
    new_string = new_string.replace("[","")
    new_string = new_string.replace("]","")
    new_string = new_string.replace("'","")
    new_string = new_string.replace('"',"")
    return new_string.split(",")
```

```python
test = df['imgs'].apply(fix_imgs)
```

```python
df['imgs'] = df['imgs'].apply(fix_imgs)
```
