```python
import json
import pandas as pd
import ast
```

```python
df = pd.read_json('complete.json')
```

```python
def fix_imgs(string):
  new_string = string.strip("[]'")
  return [new_string]
```

```python
def fix_reviews(string):
  new_string = string.strip("[]'")
  return [new_string]
```

```python
def fix_specs(string):
  dictionary = ast.literal_eval(string)
  return dictionary
```

```python
df['imgs'] = df['imgs'].apply(fix_imgs)
```

```python
df['reviews'] = df['reviews'].apply(fix_reviews)
```

```python
df['specifications'] = df['specifications'].apply(fix_specs)
```
