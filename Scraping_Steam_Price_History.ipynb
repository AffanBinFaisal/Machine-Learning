```python
import requests
from bs4 import BeautifulSoup
import re
```

```python
url = "https://steamcommunity.com/market/listings/730/Danger%20Zone%20Case"
```

```python
response = requests.get(url)
page_source = response.text
soup = BeautifulSoup(page_source, "html.parser")
scripts = soup.find_all("script")

prices = []

for script in scripts:
    script_text = str(script)
    if "line1" in script_text:
        match = re.search(r'line1\s*=\s*(\[.*?\]);', script_text, re.DOTALL)
        if match:
            line1_list = match.group(1)
            prices = eval(line1_list)  # Use eval to convert the JavaScript array to a Python list
```

```python
prices
```

```python
import matplotlib.pyplot as plt
```

```python
x = []
y = []

for price in prices:
  x.append(price[0])
  y.append(price[1])
```

```python
plt.plot(x, y)
plt.xlabel("Days")
plt.ylabel("Price")
plt.title("Price History")
plt.show()
```
