```python
def mse_gradient(weights, y_pred, X, y_true):
  m = len(y_true)
  errors = y_pred - y_true
  return (1/m) * np.dot(X.T, errors)
```

```python
def gradient_descent(X, y_true, learning_rate, epochs=1000):
  weights = np.zeros(X.shape[1])

  for epoch in range(epochs):
    y_pred = linear_regression_vector_prediction(weights, X)
    gradient = mse_gradient(weights, y_pred, X, y_true)
    weights = weights - learning_rate * gradient
  return weights
```

```python
import numpy as np
from sklearn import datasets

def linear_regression_prediction(weights, X):
    return np.dot(X, weights)

def mean_squared_error(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

def gradient_descent(X, y, learning_rate, epochs=1000):
    m, n = X.shape
    weights = np.zeros(n)

    for epoch in range(epochs):
        y_pred = linear_regression_prediction(weights, X)
        gradient = (2/m) * X.T.dot(y_pred - y)
        weights = weights - learning_rate * gradient

    return weights

# Load California housing dataset
california_housing = datasets.fetch_california_housing()
X = california_housing.data
y = california_housing.target

# Add a bias term to the feature matrix
X_bias = np.c_[np.ones((X.shape[0], 1)), X]
```

```python
learning_rate = 0.00000001
weights = gradient_descent(X_bias, y, learning_rate=learning_rate, epochs=100000)
prediction = linear_regression_prediction(weights, X_bias[100])
mse = mean_squared_error(y, linear_regression_prediction(weights, X_bias))

print("Prediction:", prediction)
print("True Value:", y[100])
print("Mean Squared Error:", mse)
```
